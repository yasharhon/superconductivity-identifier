## Help functions

```python
def print_labels(labels):
    for label in labels:
        print("\"" + label + "\",")

def power_func(x, a, nu, b):
    return a*np.power(x, nu) + b
        
def make_pair_plot(data_obj, plot_vars):
    sns.pairplot(data = data_obj,
                 vars = plot_vars)

    plt.suptitle("PairPlot of Data",fontsize=18)

    plt.tight_layout()

def make_correlation_matrix(data_obj, labels, thres):
    plt.figure(figsize = (16, 12))
    sns.heatmap(data = data_obj[labels].corr(),
                vmin = -thres,
                vmax = thres,
                annot=True,
                cmap="bone",
                linewidths=0.1,
                fmt=".2f",
                linecolor="gray")
    
    plt.title("Correlation Matrix")
    
    plt.tight_layout()

def plot_vars_target(data_obj, plot_vars, target):
    for var in plot_vars:
        fig = plt.figure(figsize = FIGSIZE_GRAPH)
        ax = fig.add_subplot()

        var_data = data_obj.get(var)
        order_keys = np.argsort(var_data)

        ax.scatter(var_data[order_keys], data_obj.get(target)[order_keys], s = 1)
```

## Study of polynomial coefficient

```python
#good_labels = unique_labels[:8]
all_labels = data_obj.columns.values[:-1].tolist()
```

```python
exps = np.ones(len(all_labels))
i = 0

for label in all_labels:
    print(label)
    print(data_obj[label].values.shape)
    
    lin_poly = nppoly.polynomial.Polynomial.fit(data_obj[label].values, data_obj[TARGET].values, 1)
    y_lin = lin_poly(data_obj[label].values)
    r2_lin = skmet.r2_score(data_obj[TARGET].values, y_lin)*100
    print("Linear R2 score: ", r2_lin)
    
    fig = plt.figure()
    ax = fig.add_subplot()
    
    sorted_keys = np.argsort(data_obj[label].values)

    ax.scatter(data_obj[label].values, data_obj[TARGET].values, s = 1)
    ax.plot(data_obj[label].values[sorted_keys], y_lin[sorted_keys], color = 'g')
    
    try:
        popt, pcov = spopt.curve_fit(power_func, data_obj[label].values, data_obj[TARGET].values)
        y_power = power_func(data_obj[label].values, popt[0], popt[1], popt[2])
        print("Optimal exponent: ", popt[1])
        r2_curve = skmet.r2_score(data_obj[TARGET].values, y_power)*100
        print("Curve R2 score: ", r2_curve)
        
        ax.plot(data_obj[label].values[sorted_keys], y_power[sorted_keys], color = 'r')
    
        if np.abs(r2_curve - r2_lin) > 2:
            exps[i] = popt[1]
    except:
        pass
    
    plt.show()
    i += 1
    
print(np.array2string(exps, precision = 5, max_line_width = 3, suppress_small = True, separator = ","))
```
